## Uniqueness & keys

Questions
- Are there any primary keys?
    - If so, what are they?
- Are there any duplicate records?

```python
job_data.info()
```

```python
# How many unique values exist
job_data["job_id"].nunique()
```

```python
# How many duplicate values exists
job_data.duplicated(subset=["job_id"]).sum()
```

### Observations

- job_id might be a primary key
- There are 80 unique values in the job_id column
- There are 0 duplicated values in the job_id column
- Because there are 80 unique values in the job_id column and there are 80 rows in the data set, this would be a valid primary key
-   The .duplicated() check also confirms this, returning 0 for the number of duplicated values

Questions
